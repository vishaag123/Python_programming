```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
        
class Linkedlist:
    def __init__(self):
        self.head=None
        
    def append(self,value):
        newnode=Node(value)
        if self.head==None:
            self.head=newnode
        else:
            curnode=self.head
            while curnode.next:
                curnode=curnode.next
            curnode.next=newnode
            
    def print(self):
        curnode=self.head
        while curnode:
            print(curnode.value)
            curnode=curnode.next
            
            
    def insert(self,value,ind):
        newnode=Node(value)
        if ind==0:
            newnode.next=self.head
            self.head=newnode
        else:
            count=0
            curnode=self.head
            while curnode.next and count<ind-1:
                count+=1
                curnode=curnode.next
            newnode.next=curnode.next
            curnode.next=newnode
            
            
    def search(self,value):
        curnode=self.head
        while curnode:
            if curnode.value==value:
                return "Yes"
            curnode=curnode.next
        return "No"
    
    def delete(self,value):
        if self.search(value)=="No":
            print("Value not found")
        else:
            curnode=self.head
            prev=self.head
            while curnode.value!=value:
                prev=curnode
                curnode=curnode.next
            if curnode==self.head:
                nxtnode=curnode.next
                curnode.next=None
                self.head=nxtnode
            else:
                prev.next=curnode.next
                curnode.next=None
                
                
    def reverse(self):
        prev=None
        cur=self.head
        while cur:
            nxt=cur.next #temporary variable to store next of current
            cur.next=prev
            prev=cur
            cur=nxt
        self.head=prev
        
    def remove_duplicate(self):
        d={}
        cur=self.head
        prev=None
        while cur:
            if cur.value not in d:
                d[cur.value]=1
                prev=cur
                cur=cur.next
            else:
                prev.next=cur.next
                cur.next=None
                cur=prev.next
                
                
    def rotate(self,k):
        cur=self.head
        while cur.value!=k and cur:
            cur=cur.next
        p=cur
        q=cur.next
        c=cur.next
        while c.next:
            c=c.next
        p.next=None
        c.next=self.head
        self.head=q
            
            
            
            
llist1=Linkedlist()
llist1.append(2)
llist1.append(6)
llist1.append(8)
# llist1.insert(10,0)
# llist1.insert(12,2)
# llist1.insert(14,1)
# llist1.append(6)
# llist1.append(8)
# llist1.insert(12,2)
# llist1.insert(14,1)
# llist1.remove_duplicate()
# llist1.rotate(12)

# llist1.reverse()

# llist1.delete(10)
# print(llist1.search(9))
llist1.print()
        
```

```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None
        self.prev=None
        
        
class Dlinkedlist:
    def __init__(self):
        self.head=None
        
    def append(self,value):
        newnode=Node(value)
        if self.head is None:
            self.head=newnode
        else:
            cur=self.head
            while cur.next:
                cur=cur.next
            cur.next=newnode
            newnode.prev=cur
            
    def prepend(self,value):
        newnode=Node(value)
        if self.head is None:
            self.head=newnode
        else:
            newnode.next=self.head
            self.head.prev=newnode
            self.head=newnode
            
    def insert(self,value,ind):
        newnode=Node(value)
        if ind==0:
            self.prepend(value)
        else:
            count=0
            cur=self.head
            while count<ind-1:
                cur=cur.next
                count+=1
            nxt=cur.next
            if nxt is not None:
                cur.next=newnode
                nxt.prev=newnode
                newnode.prev=cur
                newnode.next=nxt
            else:
                cur.next=newnode
                newnode.prev=cur
                
                
    def delete(self,key):
        cur=self.head
        prv=None
        while cur.value!=key and cur.next!=None:
            prv=cur
            cur=cur.next
        if cur.value!=key and cur.next==None:
            print("key is not in list")
        elif cur.value==key and cur.next==None:
            prv.next=None
            cur.prev=None
        elif cur.value==key and cur==self.head:
            nxt=cur.next
            cur.next=None
            nxt.prev=None
            self.head=nxt
        else:
            nxt=cur.next
            cur.next=None
            cur.prev=None
            prv.next=nxt
            nxt.prev=prv
            
            
    def reverse(self):
        cur=self.head
        tmp=None
        while cur:
            tmp=cur.prev   
            cur.prev=cur.next # Focus only on current (Alter next to prev & prev to next)
            cur.next=tmp      # Focus only on current (Alter next to prev & prev to next)
            cur=cur.prev      # To move forward in loop update cur according to original one
        if tmp:
            self.head=tmp.prev
        
            
            
                
            
            
            
    def print(self):
        cur=self.head
        while cur:
            print(cur.value)
            cur=cur.next
            
            
dllist1=Dlinkedlist()
dllist1.append(2)
dllist1.append(6)
dllist1.append(8)
dllist1.prepend(3)
dllist1.insert(5,3)
dllist1.reverse()
# dllist1.delete(9)

dllist1.print()
            
            
```
